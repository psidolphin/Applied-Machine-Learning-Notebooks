**Problem 5** Write a Python function called `spell_backwards` that takes a list of strings as an input and returns another list that contains all elements of the original list written in reverse and capitalized (only the first character must be capitalized, all others must be lowercase). For example, if the input is a list of strings `["Alexey", "John", "Anna"]` then the output must be `["Yexela", "Nhoj", "Anna"]`. Note that "Anna" is a *palindrome* which is a word, phrase, or sentence that reads the same backward or forward.

Demonstrate that the function works as expected by feeding it the string `["Alexey", "John", "Anna"]`.

```
namelist = ["Alexey", "John", "Anna"]
def spell_backwards(name_list):
  newlist = []
  for i in name_list:
    string = i[::-1]
    string = string.capitalize()
    newlist.append(string)
  return newlist
spell_backwards(namelist)
```

```
a = 'abracadabra'
b = 'palindrome'
def string_transform(string):
  if len(string) % 2 != 0:
    print(string.upper())
  else:
    return string.split()
string_transform(a)
string_transform(b)
```
