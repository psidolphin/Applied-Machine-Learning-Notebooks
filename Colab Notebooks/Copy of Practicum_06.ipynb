### Practicum 6: Decision Trees.

```
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris

iris_data=load_iris()
iris = pd.DataFrame(data=iris_data['data'], columns=[col[:-5] for col in iris_data['feature_names']])
iris['target']=iris_data['target']
cols=['petal length', 'petal width', 'target']
iris=iris[cols]
```

```
iris.head()
```

```
iris.info()
```

```
from sklearn.tree import DecisionTreeClassifier
tree = DecisionTreeClassifier(max_depth=2,random_state=42)
```

```
X=iris.loc[:, ['petal length', 'petal width']].values
y =iris.loc[:,'target'].values
```

```
tree.fit(X,y)
```

```
from sklearn.tree import export_graphviz

export_graphviz(
    tree,
    out_file='iris.dot',
    feature_names=['petal length', 'petal width'],
    class_names = ['setosa', 'versicolor', 'virginica'],
    rounded=True,
    filled=True,
)
```
