**Problem 1** 
1. Set the NumPy random seed equal to a value of 2021 (the current year). (Don't forget to import the NumPy library first!)
2. Then create a 3 by 5 (rows$\times$columns -- from now on this is going to be our standard convention) NumPy array of random floating point numbers uniformly distributed from 30 to 50 (inclusive). Call this array `rand_floats` and print it to the screen. 
3. Next, set the same random seed again and create a 3 by 5 NumPy array of random values pulled from the normal distribution of mean 40 and standard deviation 10 (use the built-in help system if you are not sure about the syntax). Call this array `rand_norm` and print it to the screen.
4. Measure the average time it takes to add these two arrays using the `+` operator (do it is a separate code cell).
5. Add the arrays using the `+` operator and store  the result in a new array called `floats_plus_norm`.
5. Print `floats_plus_norm` to the screen.

```
import numpy as np
np.random.seed(2021)
rand_floats = np.random.uniform(30,50, size=(3,5))
np.random.seed(2021)
rand_norm = np.random.normal(loc=40.0,scale=10.0,size=(3,5))
floats_plus_norm = rand_floats + rand_norm
print(floats_plus_norm)
```

**Problem 3** 
1. Use an appropriate NumPy function to create an increasing sequence of consecutive positive integers from 0 to 19 (inclusive). Store this sequence in a variable called `pos` and print this variable to the screen.
2. In a similar fashion, create a decreasing sequence of consecutive negative integers from 0 to -19 (inclusive). Store this sequence in a variable called `neg` and print this variable to the screen. 
3. To check that your sequences are correct, add `pos` and `neg` together -- the result must be an array of 20 zeros.
4. Concatenate the first 4 elements of `pos` with the last 3 elements of `neg` (in this order) and print the result to the screen.
5. Reshape both `pos` and `neg` into a 5 by 4 matrices called `pos_mat` and `neg_mat`. Print the matrices to the screen.
6. Stack matrices row-wise placing `pos_mat` on top. The result must be a new 10 by 4 matrix. Print this matrix to the screen.
7. Stack matrices column-wise placing `neg_mat` at the leftmost position. The result must be a new 5 by 8 matrix. Print this matrix to the screen.

```
import numpy as np
pos = np.arange(0,20)
print(pos)
neg = np.arange(0,-20,-1)
print(neg)
print(pos + neg)
print('\n')
np.concatenate((pos[0:5],neg[-3:]))
pos_mat = pos.reshape((5,4))
neg_mat = neg.reshape((5,4))
print('\n')
print(pos_mat)
print('\n')
print(neg_mat)
print('\n')
print(np.vstack((pos_mat,neg_mat)))
print('\n')
print(np.hstack((pos_mat,neg_mat)))
```
